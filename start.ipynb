```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('financial_regression.csv')

display(df.head())

df.info()
```

```python
# Convertir la columna 'date' a tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Ordenar los datos del más antiguo al más reciente
df = df.sort_values(by='date').reset_index(drop=True)

print(f"Rango de fechas: desde {df['date'].min().date()} hasta {df['date'].max().date()}")
```

Como no hay PIB (GDP) de los primeros 3 meses y hay datos de sobra, se eliminan los primeros 3 meses

```python
# Eliminar todos los datos anteriores al 2010-04-01
df = df[df['date'] >= '2010-04-01'].reset_index(drop=True)

# Comprobar el nuevo rango de fechas y la cantidad de filas
print(f"Nuevo rango de fechas: desde {df['date'].min().date()} hasta {df['date'].max().date()}")
print(f"Total de filas tras el recorte: {len(df)}")
```

```python
# Lista de columnas macroeconómicas a rellenar
columnas_macro = ['GDP', 'CPI', 'us_rates_%']

# Aplicar forward fill solo a estas columnas
df[columnas_macro] = df[columnas_macro].ffill()

print("Valores nulos tras el relleno:")
print(df[columnas_macro].isna().sum())
```
